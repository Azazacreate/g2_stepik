```python
# получить 4-цифрыПоследние_картКредитных
def get__numbersLast(num):
    return int(str(num)[-4:])

df['CC Number'].apply(get__numbersLast)
```

```python
df['numbersLast'] = df['CC Number'].apply(get__numbersLast)
df.head()
```

```python
def yelp(price):
    if price < 13.3475:
        return '$'
    elif 13.3475 <= price < 17.795:
        return '$$'
    else:
        return '$$$'

df['total_bill'].apply(yelp) + df['total_bill'].apply(str)
```
